# GANs Image Generation

In this notebook, we will explore how Generative Adversarial Networks (GANs) generate images. We will use a pretrained GAN model (BigGAN) to generate images from random noise.

```python
# Install + import required NLTK package for BigGAN capabilities.
!pip install nltk

import nltk
nltk.download('wordnet')

print("NLTK and WordNet installed successfully!")
```

```python
import torch 
from pytorch_pretrained_biggan import (BigGAN, one_hot_from_names, truncated_noise_sample,
                                       save_as_images, display_in_terminal)
from PIL import Image  
import os

# Load pre-trained GAN model for effective 128x128 pixels image generation.
model = BigGAN.from_pretrained('biggan-deep-128')

truncation = 0.2 #Value for determining quality/diversity of images (Lower the value -> better resolution)
#Map noise vectors to corresponding image type.
latent_vector = torch.randn(3, 128) 
class_vector = one_hot_from_names(['soap bubble', 'coffee', 'mushroom'], batch_size=3)

# All in tensors
noise_vector = latent_vector  # Use latent_vector as the noise vector
class_vector = torch.from_numpy(class_vector)

# Generate an image
with torch.no_grad():
    output = model(noise_vector, class_vector, truncation)

# Save results as png images in the specified folder

# Create the directory if it doesn't exist
output_dir = "images"
os.makedirs(output_dir, exist_ok=True)

# Save each image in the output tensor
# Save each image in the output tensor
for i, img in enumerate(output):
    img = img.squeeze().cpu().numpy()
    img = (img * 255).astype('uint8')  # Convert to uint8 format
    img = Image.fromarray(img.transpose(1, 2, 0))  # Convert to HWC format for saving
    img.save(os.path.join(output_dir, f"image_{i}.png"))
```

```python
# Experiment with different latent vectors
latent_vector = torch.randn(3, 128)  # Change this to explore different random vectors
with torch.no_grad():
    generated_images = model(latent_vector, class_vector, truncation)

# Display multiple images using matplotlib
import matplotlib.pyplot as plt
plt.figure(figsize=(15, 5))

for i, img in enumerate(generated_images):
    # Process the image data correctly
    img = img.cpu().numpy()  # Convert to numpy array
    img = (img * 255).astype('uint8')  # Scale to 0-255 range
    img = img.transpose(1, 2, 0)  # Convert from CHW to HWC format
    
    # Display in subplot
    plt.subplot(1, 3, i+1)
    plt.imshow(img)
    plt.title(f"Image {i+1}")
    plt.axis('off')

plt.show()
```
